```python
dfcopy = df.copy()

dfcopy = dfcopy.replace({'*': '0', 99999: 1, '99999': '1'})
```

```python
for col in dfcopy.columns:
  print(col)
  print(dfcopy[col].nunique())
  print(set(dfcopy[col].values))
  print()
  print()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
# Convert labels to numerical values using LabelEncoder
label_encoder = LabelEncoder()
for col in dfcopy.columns:
  if col in ['protocol_type', "service", "flag", 'type_of_attack']:
    dfcopy[col] = label_encoder.fit_transform(dfcopy[col])
  else:
    dfcopy[col] = dfcopy[col].astype('float64')
```

```python
dfcopy = dfcopy.dropna()
y = dfcopy['type_of_attack']
X = dfcopy.drop(['type_of_attack'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
models = [
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('SVM', SVC()),
    ('XGBoost', xgb.XGBClassifier()),
    ('K-Nearest Neighbors', KNeighborsClassifier())
]
```

```python
accs = {}

for model_name, model in models:
    print(f"Training and evaluating {model_name}...")
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, predictions)
    report = classification_report(y_test, predictions)

    print(f"Model: {model_name}")
    print("Accuracy:", accuracy)
    accs[model_name] = accuracy
    print("Classification Report:\n", report)
    print("="*50)
```
